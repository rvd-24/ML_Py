```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year,df.PCI,color='red',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']].values,df.PCI)
```

```python
reg.predict([[1970]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d=pd.read_csv('canada_per_capita_income.csv')
d
```

```python
p=reg.predict(np.array(d.year).reshape(-1,1))
p
```

```python
df['Predicted Per Capita Income']=p
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df.year,df.PCI,color='red',marker='*')
plt.plot(df.year,reg.predict(df[['PCI']].values),color='blue')
plt.show()
```

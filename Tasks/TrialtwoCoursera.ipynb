```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Coursera.csv")
df.head()
```

```python
df=df.drop(['Course URL','Course Description','University'],axis='columns')
df.head()
```

```python
df['Difficulty Level'].unique()
```

```python
df=df[df['Difficulty Level'] != 'Not Calibrated']
df['Difficulty Level'].unique()
```

```python
dummies=pd.get_dummies(df['Difficulty Level'])
df=pd.concat([df.drop(['Difficulty Level'],axis='columns'),dummies],axis='columns')
```

```python
df.head()
```

```python
df['Skills']=df['Skills'].str.lower().str.split('  ')
df.head()
```

```python
columns=['Course Name','Skills']
df['metadata']=df[columns].to_dict(orient='records')
```

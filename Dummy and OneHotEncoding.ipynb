```python
import pandas as pd
df=pd.read_csv('hotencodinghomeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#Dropping town column and another town as it can be derived from other two towns in order to avoid dummy trap
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
#2800sq ft. in Robinsville
model.predict([[2800,0,1]])
```

```python
#3400sq.ft. in West Windsor
model.predict([[3400,0,0]])
```

```python
#Compares Predicted and Actual Values(1 for Perfect)
model.score(x,y)
```
